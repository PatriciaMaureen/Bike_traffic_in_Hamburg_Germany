```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/yearly_bike_data_daily.csv')
```

```python
df
```

```python
df['Asset ID'].value_counts()
```

```python
#make all column names lower case
df.columns=df.columns.str.lower()

#replace blank spaces
df.columns = df.columns.str.replace(' ','_')
df
```

```python
# We can drop the resulttime, since it just tells us when the data was stored.
df = df.drop(['resulttime'], axis=1)
```

```python
# Let's see if we have any NaN values.
# Looks like we do not have null vlaues. 

df.isna().sum()
```

```python
# Let's check fpr duplicates
# Fortunately there are no duplicates
df.duplicated().sum()
```

```python
# Then we can check if all the columns have the right data type 

df.dtypes
```

```python
# We can change long, lat to numbers, and we can change pehnomenontime to datetime
# Before we can change pehnomenontime we have to check what date to keep 2024-05-17T22:00:00Z/2024-05-18T21:59:59Z
# The column holds the start and end point of the measurement. Since we start at 10 in evening and we want to look at daily data, I will keep the day were 22 hours ofd bike traffic are measured.

df['phenomenontime'] = df['phenomenontime'].apply(lambda x: x.split('/')[-1])
df
```

```python
df['latitude'] =  pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] =  pd.to_numeric(df['longitude'], errors='coerce')

df['phenomenontime'] = pd.to_datetime(df['phenomenontime'])

df.dtypes
```

```python
df.isna().sum()
```

```python
# Now that we have the date column we can create a column for the day, the weekday and the month
# year does not really make sense because we got one whole year of data starting in 2023 and ending in 2024

df['month'] = df['phenomenontime'].dt.month
df['day'] = df['phenomenontime'].dt.day
df['weekday'] = df['phenomenontime'].dt.weekday

# If we want to merge more data based on the date, it would be nice to just have the data in the phenomenontime	 column

df['phenomenontime'] = df['phenomenontime'].dt.date
```

```python
# If I just want to keep the date, as executed before, it is stored as type object so I have to convert it back into datetime format.
df['phenomenontime'] = pd.to_datetime(df['phenomenontime'])
```

```python
df.dtypes
```

```python
# Let's merge the data about the state_holidays and school_holidays with this. 
# I got the information here and stored it in a .csv file manually: https://www.hamburg.de/bsb/ferien/

df_holidays = pd.read_csv('data/holidays_hamburg.csv')
df_holidays
```

```python
df_holidays['date'] = pd.to_datetime(df_holidays['date'], format='%d.%m.%y')
df_holidays
```

```python
# Now we can merge the two dataframes together
```

```python
df = pd.merge(left = df, right = df_holidays, how = 'left', left_on = "phenomenontime", right_on= "date")
df
```

```python
# Then we can drop the date column, because now we have it twice in there

df = df.drop(['date'], axis=1)
```

```python
df.isna().sum()
```

```python
# Let's now get the weather data and merge it, too.

df_weather = pd.read_csv('data/weather_data_daily.csv')
df_weather
```

```python
df_weather.dtypes
```

```python
# First we have to change the date column to datetime format, otherwise we can not merge the two dataframes based on this column
df_weather['date'] = pd.to_datetime(df_weather['date'], format='%Y%m%d')
df_weather
```

```python
df = pd.merge(left = df, right = df_weather, how = 'left', left_on = "phenomenontime", right_on= "date")
df
```

```python
# Then we can drop the date column, because now we have it twice in there

df = df.drop(['date'], axis=1)
```

```python
# Let's use a boxplot graph to see if there are any outliers in the columns

import seaborn as sns
import matplotlib.pyplot as plt

for column in df_num:
    sns.boxplot(x=df_num[column])
    plt.show()
```
