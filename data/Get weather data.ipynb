```python
# DWD (DEutscher WEtterdienst)
# https://www.dwd.de/DE/leistungen/klimadatendeutschland/klarchivtagmonat.html;jsessionid=CA4E0930EDAC2FBA834C0FB5CFE23503.live21061?nn=16102

import pandas as pd

data_weather = pd.read_csv('produkt_klima_tag_20221124_20240526_01975.txt', sep = ';') 
data_weather
```

```python
# The defitnition of the column names can be found here: 
# https://www.dwd.de/DE/leistungen/klimadatendeutschland/beschreibung_tagesmonatswerte.html;jsessionid=F045DF84F28AF665D291AB58ACDDAFA1.live21061?nn=16102&lsbId=526270

# make all column names lower case
data_weather.columns=data_weather.columns.str.lower()

# I got an error mesage when I tried to drop some of the columns, so I wanted to check the column names for blank spaces.
data_weather.columns = data_weather.columns.str.replace(' ','_')
data_weather
```

```python
# Let's drop the columns we do not need
data_weather = data_weather.drop(['stations_id','qn_3','qn_4','_sdk','shk_tag','__nm','_vpm','__pm','_upm','_tgk','eor'], axis=1)
data_weather
```

```python
# Let's rename the column names 

data_weather = data_weather.rename(columns={'mess_datum':'date','__fx':'max_wind_speed','__fm':'mean_wind_speed','_rsk':'precipitation_height', 'rskf':'precipitation_catgeory','_tmk':'mean_temperature','_txk':'max_temperature','_tnk':'min_temperature' })
data_weather
```

```python
# Precipitation_category description:

# 0 no precipitation 
# 1 only rain (in historical data before 01.01.1979)
# 4 Form of precipitation not known, although precipitation reported
# 6 only rain; liquid precipitation with automatic measurement
# 7 only snow; solid precipitation with automatic measurement
# 8 Rain and snow and/or sleet; liquid and solid precipitation with automatic measurement
# 9 Incorrect identification; missing value or precipitation form cannot be determined during automatic measurement

```
